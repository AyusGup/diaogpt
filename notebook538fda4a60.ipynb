```python
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments, DataCollatorForLanguageModeling
from torch.utils.data import Dataset, DataLoader
import torch
```

```python
# Load fine-tuned model and tokenizer
model = AutoModelForCausalLM.from_pretrained("/kaggle/input/dialogpt-fine-tunedcustomer-service/pytorch/default/1/fine-tuned-dialoGPT")
tokenizer = AutoTokenizer.from_pretrained("/kaggle/input/dialogpt-fine-tunedcustomer-service/pytorch/default/1/fine-tuned-dialoGPT")
```

```python
# Test the model
input_text = "What things are there in a Dinosaur Museum? "
input_ids = tokenizer.encode(input_text + tokenizer.eos_token, return_tensors="pt")
output_ids = model.generate(input_ids, max_length=100)

response = tokenizer.decode(output_ids[:, input_ids.shape[-1]:][0], skip_special_tokens=True)
print(f"Bot: {response}")
```
